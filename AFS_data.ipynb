```python
pd_afs.head()
```

```python
columns_to_drop2 = [
    "rd_blends_fr",
    "intersection_directions_fr",
    "bd_blends_fr",
    "ev_pricing_fr",
    "ev_charging_units"
]
pd_afs = pd_afs.drop(columns=columns_to_drop2)
```

```python
pd_afs.head()
```

```python
pd_afs = pd_afs[pd_afs["fuel_type_code"].isin(["ELEC","BD"])]

print("Filtered AFS shape:", pd_afs.shape)
pd_afs["fuel_type_code"].value_counts()
```

```python
pd_afs["zip"] = (
    pd_afs["zip"]
    .astype(str)
    .str.replace(r"\.0$", "", regex=True)
    .str.zfill(5)
)
```

```python
unique_zips = pd_afs['zip'].drop_duplicates().reset_index(drop=True)
print(unique_zips)
```

```python
zip_county = "/content/Zip Code to County Spreadsheet export 2025-12-02 22-31-09.zip"
zip = pd.read_csv(zip_county)
zip.head()
```

```python
zip_il = zip[zip['USPS_ZIP_PREF_STATE'] == 'IL']
display(zip_il.head())
```

```python
il_zip_county_df = zip_il.copy()
display(il_zip_county_df.head())
```

```python
il_zip_county_df['ZIP'].nunique()
```

```python
il_zip_county_df['County Clean'].nunique()
```

```python
counties_aqi = [
    "Adams", "Champaign", "Clark", "Cook", "DuPage", "Effingham", "Hamilton",
    "Jersey", "Jo Daviess", "Kane", "Lake", "Macon", "Macoupin", "Madison",
    "McHenry", "McLean", "Peoria", "Randolph", "Rock Island", "Saint Clair",
    "Sangamon", "Will", "Winnebago"
]


filtered_df = il_zip_county_df[il_zip_county_df['County Clean'].isin(counties_aqi)]

unique_zips_df = filtered_df[['ZIP']].drop_duplicates().reset_index(drop=True)

print(unique_zips_df)
```

```python
set_of_afs_zips = set(unique_zips[unique_zips != '00nan'].astype(str))
set_of_aqi_zips = set(unique_zips_df['ZIP'].astype(str))

shared_unique_zips = list(set_of_afs_zips.intersection(set_of_aqi_zips))

print("Shared unique ZIP codes found:")
print(shared_unique_zips)
print(f"\nTotal number of shared unique ZIP codes: {len(shared_unique_zips)}")
```

```python
pd_afs_filtered = pd_afs[pd_afs['zip'].isin(shared_unique_zips)]
display(pd_afs_filtered.head())
```

```python
columns_to_remove = ['x', 'y', 'county_x', 'county_y']
pd_afs_filtered = pd_afs_filtered.drop(columns=columns_to_remove)
display(pd_afs_filtered.head())
```

```python
pd_afs_filtered = pd_afs_filtered[pd_afs_filtered["fuel_type_code"].isin(["ELEC", "BD"])]

print("Filtered AFS shape:", pd_afs_filtered.shape)
pd_afs_filtered["fuel_type_code"].value_counts()
```

```python
zip_count = (
    pd_afs_filtered.groupby(["zip", "fuel_type_code"])
       .size()
       .reset_index(name="count")
)

zip_count.head()
```

```python
zip_to_county_mapping = il_zip_county_df[['ZIP', 'County Clean']].drop_duplicates(subset=['ZIP']).copy()
zip_to_county_mapping.rename(columns={'ZIP': 'zip', 'County Clean': 'County'}, inplace=True)


pd_afs_filtered['zip'] = pd_afs_filtered['zip'].astype(str)
zip_to_county_mapping['zip'] = zip_to_county_mapping['zip'].astype(str)


pd_afs_filtered = pd.merge(
    pd_afs_filtered,
    zip_to_county_mapping,
    on='zip',
    how='left'
)

print("DataFrame after adding county information:")
display(pd_afs_filtered.head())
```

```python
county_count = (
    pd_afs_filtered.groupby(["County_x", "fuel_type_code"])
       .size()
       .reset_index(name="count")
)

county_count
```

```python
elec_county_counts = county_count[county_count['fuel_type_code'] == 'ELEC'][['County_x', 'count']]
display(elec_county_counts.head())
```

```python
bd_county_counts = county_count[county_count['fuel_type_code'] == 'BD'][['County_x', 'count']]
display(bd_county_counts.head())
```
